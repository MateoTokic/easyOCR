**Import libraries**

```python
import easyocr
import matplotlib.pyplot as plt
import cv2
```

```python
# Initialize the reader for English
reader = easyocr.Reader(['hr'])

# Read the image
image_path = '/content/drive/MyDrive/proizvodi/petit_keksi.jpg'
image = cv2.imread(image_path)
```

```python
# Use EasyOCR to detect and recognize text
result = reader.readtext(image)

# Print the recognized text
for (bbox, text, prob) in result:
    print(f'Detected text: {text} (Confidence: {prob})')
```

```python
# Draw bounding boxes on the image
for (bbox, text, prob) in result:
    top_left = tuple([int(v) for v in bbox[0]])
    bottom_right = tuple([int(v) for v in bbox[2]])
    cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)

# Display the image with bounding boxes
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
